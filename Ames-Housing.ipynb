```python
import pandas as pd
```

```python
# Question 1:
# Load the Ames Housing dataset
ames_df = pd.read_csv('AmesHousing.csv')
```

```python
# Display first five rows of the DataFrame
ames_df.head()
```

```python
# Question 2:
# Highest priced home
highest_price = ames_df.loc[ames_df['SalePrice'].idxmax()]

# Lowest priced home
lowest_price = ames_df.loc[ames_df['SalePrice'].idxmin()]

print("Highest Priced Home:")
print(highest_price)

print("\nLowest Priced Home:")
print(lowest_price)
```

```python
# Question 3:
# Count distinct categories of Lot Size
lot_size_categories = ames_df['Lot Config'].nunique()
print("Number of categories for Lot Size:", lot_size_categories)
```

```python
# Question 4: 
# Calculate range of years for house listings
min_year = ames_df['Year Built'].min()
max_year = ames_df['Year Built'].max()
print("Range of years for house listings:", min_year, "to", max_year)
```

```python
# Question 5:
# Total number of homes sold in Ames
total_num_sold = ames_df.shape[0]
print("Total number of homes sold in Ames:", total_num_sold)
```

```python
# Question 6:
# Count nulls in Lot Frontage column
count_nulls = ames_df['Lot Frontage'].isnull().sum()
print("Number of nulls or NA in Lot Frontage column:", count_nulls)
```

```python
# Question 7:
# Count non-nulls and nulls in Alley column
count_alley = ames_df['Alley'].notnull().sum()
nulls_alley = ames_df['Alley'].isnull().sum()
print("Number of items with a value in Alley column:", count_alley)
print("Number of nulls or NA in Alley column:", nulls_alley)
```

```python
# Question 8:
# Filter homes in Gilbert with Yr Sold in 2009 or 2010
gilbert_homes = ames_df[(ames_df['Neighborhood'] == 'Gilbert') & (ames_df['Yr Sold'].isin([2009, 2010]))]
print("Homes in Gilbert with Yr Sold in 2009 or 2010:")
print(gilbert_homes)
```

```python
# Question 9:
# count of homes sold in Gilbert with Yr Sold being 2010
count_homes = ames_df[(ames_df['Neighborhood'] == 'Gilbert') & (ames_df['Yr Sold'] == 2010)].shape[0]
print("Homes in Gilbert with Yr Sold in 2010:", count_homes)
```

```python
# Question 10:
# Access value at third row, fifth column (indexing starts at 0)
my_value = ames_df.iloc[2, 4]
print("Value at third row, fifth column:", my_value)
```

```python
# Question 11:
# Format SalePrice column with $ and commas
ames_df['Formatted SalePrice'] = ames_df['SalePrice'].apply(lambda x: '${:,.0f}'.format(x))

# Display first five rows to verify
ames_df[['SalePrice', 'Formatted SalePrice']].head()
```

```python
# Question
# Create new DataFrame with required columns
minimal_df = ames_df[['Neighborhood', 'Year Built', 'Formatted SalePrice']]

# Save DataFrame to CSV file
minimal_df.to_csv('minimal_ames_housing.csv', index=False)

# Print confirmation message
print("DataFrame saved as 'minimal_ames_housing.csv'")
```
